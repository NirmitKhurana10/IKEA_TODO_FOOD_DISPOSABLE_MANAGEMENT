```python
import pandas as pd
```

```python
df = pd.read_excel("data/data.xlsx")
```

```python
df
```

```python
df = df.dropna()
df
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Step 2: Split the 'Item' column into 'batch_code' and 'item_name'
df[['batch_code', 'item_name']] = df['Item'].str.split(' - ', n=1, expand=True)
```

```python
df
```

```python
new_cols = ['batch_code', 'item_name', 'Items per 1000 customers', 'Receipts', 'Sales', 'Quantity']
df = df[new_cols]
```

```python
df
```

```python
drop_cols = ['batch_code', 'item_name', 'Quantity']
df = df[drop_cols]
```

```python
df[['batch_code', 'Quantity']] = df[['batch_code', 'Quantity']].astype(int)
df
```

```python
df.rename(columns={'Quantity':'quantity_sold'}, inplace=True)
```

```python
df['week_number'] = 25  # Update this based on your reporting week
```

```python
df
```

```python
df.to_csv("week_25_sales_cleaned.csv", index=False)
```
