```python
import os
from supabase import create_client, Client

url: str = os.environ.get("SUPABASE_URL")
key: str = os.environ.get("SUPABASE_KEY")
supabase: Client = create_client(url, key)
```

```python
response = supabase.table("food_items").select("id, batch_code").execute()
food_items_data = response.data
```

```python
food_items_data
```

```python
import pandas as pd

food_items_df = pd.DataFrame(food_items_data)
food_items_df["batch_code"] = food_items_df["batch_code"].astype(int)  # match dtype
```

```python
sales_file = pd.read_csv("week_25_sales_cleaned.csv")
sales_df = pd.DataFrame(sales_file)
```

```python
merged_df = sales_df.merge(food_items_df, on="batch_code", how="inner")
```

```python
merged_df
```

```python
sales_records = merged_df[["week_number","id","quantity_sold"]].rename(columns={"id":"food_item_id"}).to_dict("records")
```

```python
for row in sales_records:
    supabase.table("food_sales").insert(row).execute()
```
